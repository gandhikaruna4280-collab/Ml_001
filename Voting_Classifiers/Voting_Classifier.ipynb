# Step 1: Import Required Libraries

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

# Step 2: Load and Split Data

```python
# Load the dataset
iris = load_iris()
X, y = iris.data, iris.target
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Step 3: Define Base Models

```python
# Define base models
model1 = LogisticRegression(random_state=42)
model2 = DecisionTreeClassifier(random_state=42)
model3 = SVC(probability=True, random_state=42)  # Enable probability for soft voting
```

# Step 4: Create Voting Classifiers

```python
# Hard Voting Classifier
hard_voting = VotingClassifier(
    estimators=[('lr', model1), ('dt', model2), ('svc', model3)],
    voting='hard'
)
```

```python
# Soft Voting Classifier
soft_voting = VotingClassifier(
    estimators=[('lr', model1), ('dt', model2), ('svc', model3)],
    voting='soft'
)
```

# Step 5: Train and Evaluate the Models

```python
# Train and evaluate hard voting classifier
hard_voting.fit(X_train, y_train)
y_pred_hard = hard_voting.predict(X_test)
print(f"Hard Voting Accuracy: {accuracy_score(y_test, y_pred_hard):.2f}")
```

```python
# Train and evaluate soft voting classifier
soft_voting.fit(X_train, y_train)
y_pred_soft = soft_voting.predict(X_test)
print(f"Soft Voting Accuracy: {accuracy_score(y_test, y_pred_soft):.2f}")
```

# Step 6: Compare Individual Model Performance

```python
# Evaluate individual models
for model in [model1, model2, model3]:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"{model.__class__.__name__} Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```
